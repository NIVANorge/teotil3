# Read seNorge-2018

Example of reading gridded rainfall and temperature data from Met.no's Thredds server (based on seNorge 2018).

```python
year = 2022 # From 1957 to present
nc_path = f"https://thredds.met.no//thredds/dodsC/senorge/seNorge_2018/Archive/seNorge2018_{year}.nc"
ds = xr.open_dataset(nc_path)
ds
```

```python
# Pot rainfall for day of interest
ds['rr'].sel(time="2022-07-01").plot()
```
